# Data preprocessing

```python
data.head(20)
```

```python
data['is_default'] = data['default'].apply(lambda row: 1 if row == 'yes' else 0)
```

```python
data[['default','is_default']].tail(20)
```

```python
data['is_housing'] = data['housing'].apply(lambda row: 1 if row == 'yes' else 0)
data[['housing','is_housing']].tail(20)
```

```python
data['is_loan'] = data['loan'].apply(lambda row: 1 if row == 'yes' else 0)
data[['loan', 'is_loan']].tail(20)
```

```python
data['target'] = data['y'].apply(lambda row: 1 if row == 'yes' else 0)
data[['y', 'target']].tail(200)
```

```python
M_dummies = pd.get_dummies(data['marital'], prefix = 'marital')
M_dummies.tail(20)
```

```python
pd.concat([data['marital'], M_dummies], axis=1).head(n=10)
```

```python
M_dummies.drop('marital_divorced', axis=1, inplace=True)
M_dummies.head()
```

```python
data = pd.concat([data, M_dummies], axis=1)
data.head(10)
```

```python
j_dummies = pd.get_dummies(data['job'], prefix = 'job')
j_dummies.tail(10)
```

```python
j_dummies.drop('job_unknown', axis=1, inplace=True)
```

```python
data = pd.concat([data, j_dummies], axis=1)
data.head(10)
```

### one hot encoding 

```python
e_dummies = pd.get_dummies(data['education'], prefix = 'education')
e_dummies.tail(10)
```

```python
e_dummies.drop('education_unknown', axis=1, inplace=True)
e_dummies.tail(10)
```

```python
data = pd.concat([data, e_dummies], axis=1)
data.head()
```

```python
c_dummies = pd.get_dummies(data['contact'], prefix = 'contact')
c_dummies.tail(10)
```

```python
c_dummies.drop('contact_unknown', axis=1, inplace=True)
c_dummies.tail(10)
```

```python
data = pd.concat([data, c_dummies], axis=1)
data.head(10)
```

```python
poutcome_dummies = pd.get_dummies(data['poutcome'], prefix = 'poutcome')
poutcome_dummies.tail(10)
```

```python
poutcome_dummies.drop('poutcome_unknown', axis=1, inplace=True)
poutcome_dummies.tail(10)
```

```python
data = pd.concat([data, poutcome_dummies], axis=1)
data.head(10)
```

```python
months = {'jan':1, 'feb':2, 'mar':3, 'apr':4, 'may':5, 'jun':6, 'jul':7, 'aug':8, 'sep':9, 'oct':10, 'nov':11, 'dec': 12}
data['month'] = data['month'].map(months)
data['month'].head(10)
```

```python
data[data['pdays'] == -1]['pdays'].count()
```

```python
data['was_contacted'] = data['pdays'].apply(lambda row: 0 if row == -1 else 1)
data[['pdays','was_contacted']].head(10)
```

```python
data.drop(['job', 'education', 'marital', 'default', 'housing', 'loan', 'contact', 'pdays', 'poutcome', 'y'], axis=1, inplace=True)
```

```python
data.dtypes
```

```python
data.head(20)
```

```python
X = data.drop('target', axis=1)
y = data['target'] 
```

```python
X.shape
```

```python
y.shape
```

# Data train test split

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.1, random_state = 32)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

# Model Building :

## Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
log = LogisticRegression()
```

```python
log.fit(X_train, y_train)
```

```python
y_pred = log.predict(X_test) ##Predict test data
```

```python
print("Predicted value: ", y_pred[:10])
print("Actual value: ", y_test[:10])
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_pred = y_pred, y_true = y_test)
print(f'Accuracy  {accuracy*100:.2f}%')
```

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
random_f = RandomForestClassifier()
random_f.fit(X_train, y_train)
random_fpredictions = random_f.predict(X_test)
print("Predicted value: ", random_fpredictions[:10])
print("Actual value: ", y_test[:10])
```

```python
accuracy = accuracy_score(y_pred = random_fpredictions, y_true = y_test)
print(f'Accuracy   {accuracy*100:.2f}%')
```

## XGBoost


```python
from xgboost import XGBClassifier
```

```python
xgb_model = XGBClassifier()
xgb_model.fit(X_train, y_train)
prediction_xgb = xgb_model.predict(X_test)
print(accuracy_score(y_test, prediction_xgb))
```

```python
accuracy = accuracy_score(y_pred = prediction_xgb, y_true = y_test)
print(f'Accuracy  {accuracy*100:.2f}%')
```

### Inference: We can see Random Forest Classifier model gives us highest accuracy i.e. 90.51%.
